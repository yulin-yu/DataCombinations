# Quantify Atypicality of Data Combination

```python
import pandas as pd
df_all1 = pd.read_csv('DataComb_Full.csv')
```

```python
import numpy as np
from numpy.linalg import norm
import itertools


   
def makelist(a):
    ab = a.replace(' ','')
    l = ab.split(';')
    l1 = [str(int(i)) for i in l]
    return l1
df_all1['dataset1'] = df_all1['STUDY_NUMS'].apply(makelist)

from sklearn.preprocessing import MultiLabelBinarizer

all_paper_data = df_all1[['DOI','dataset1']]
#all_paper_data['STUD_NUMS'] = all_paper_data['STUD_NUMS'].astype(str)
all_paper_data = all_paper_data.dropna()

mlb = MultiLabelBinarizer(sparse_output=True)     
all_paper_data = all_paper_data.join(
            pd.DataFrame.sparse.from_spmatrix(
                mlb.fit_transform(all_paper_data.pop('dataset1')),
                index=all_paper_data.index,
                columns=mlb.classes_))


all_paper_data1 = df_all1[['DOI','dataset1']]
df_all1.NumDatasets = df_all1['NumDatasets_original'].astype(int)
df_all2 = df_all1[df_all1['NumDatasets_original']>1].DOI.unique()

all_paper_data1 =all_paper_data1[all_paper_data1.DOI.isin(df_all2)]
all_paper_data1 = all_paper_data1.dropna(subset = ['dataset1'])
```

```python
import numpy as np
from numpy.linalg import norm
import itertools


def multidiscipline(aa):
    multi_score = []
    if len(aa)>1:
        #alldata = aa.split(';')
        for a in aa:
            multi_score.append((1/len(aa))*(1/len(aa)))
        
        #size_len = len(alldata)
        flist = list(itertools.combinations(aa, 2))
        sorted_list = [tuple(sorted(item)) for item in flist]
         
        for i in sorted_list:
            #print(str(i[0]))
            #print(str(i[1]))
            A = np.array(all_paper_data[str(i[0])])
            B = np.array(all_paper_data[str(i[1])])
            cosine = np.dot(A,B)/(norm(A)*norm(B))
            num_d = cosine*(1/len(aa))*(1/len(aa))
            
            multi_score.append(num_d)
        #print(multi_score)
        return 1-sum(multi_score)         
    else:
        return 1
```

```python
all_paper_data1_t = all_paper_data1.tail(20) # to run everything, delet
all_paper_data1_t['Atypicality_of_datasets_original_1']= all_paper_data1_t['dataset1'].apply(multidiscipline)
```
